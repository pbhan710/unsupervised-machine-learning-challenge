```python
# Import dependencies.
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## 1) Prepare the Data
1. Read myopia.csv into a Pandas DataFrame.
2. Remove the "MYOPIC" column from the dataset.
3. Standardize your dataset so that columns that contain larger values do not influence the outcome more than columns with smaller values.

```python
# Read the 'myopia.csv' file into a Pandas DataFrame.
fp = 'Data/myopia.csv'
df = pd.read_csv(fp)
```

```python
# Preview the DataFrame.
df.head()
```

```python
# Remove the 'MYOPIC' column from the dataset.
df = df.drop(["MYOPIC"], axis='columns')
```

```python
# Standardize the dataset using a scaler. Name the resulting DataFrame X.
X = StandardScaler().fit_transform(df)
```

## 2) Apply Dimensionality Reduction
1. Perform dimensionality reduction with PCA. How did the number of the features change?

2. Further reduce the dataset dimensions with t-SNE and visually inspect the results. To do this, run t-SNE on the principal components, which is the output of the PCA transformation.

```python
# Create and use a PCA model to reduce dimensions of the scaled dataset. Set PCA model’s explained variance to 90%.
pca = PCA(n_components=0.90)
```

```python
# Examine the shape of the reduced dataset for reduction in number of features.

# Reduce dimensions of the scaled dataset.
myopic_pca = pca.fit_transform(X)

# Transform PCA data to a DataFrame.
df_myopic_pca = pd.DataFrame(data=myopic_pca)
df_myopic_pca.head()
```
